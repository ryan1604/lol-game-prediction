```python
df = pd.read_parquet(Path("..") / "data" / "processed" / "matches_clean.parquet")
print(df.head().to_string())
```

```python
for col in df.columns[:-3]:
    values = sorted(df[col].dropna().unique())
    print(f"\n{col} ({len(values)} unique)")
    print(", ".join(map(str, values)))
```

```python
col_a = "blue_top_player"
col_b = "red_top_player"

a = set(df[col_a].dropna().unique())
b = set(df[col_b].dropna().unique())

print(f"Only in {col_a}:")
print(sorted(a - b))

print(f"\nOnly in {col_b}:")
print(sorted(b - a))

# print(f"\nIn both:")
# print(sorted(a & b))
```
